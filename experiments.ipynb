```python
import numpy as np
import pandas as pd
import pickle
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data
```

```python
"""Gender encoder will be used for converting string or non numeric values to numeric values but when it contains only two diffferent non numeric values then only we should use Label Encoder else for more than two ot will still assign numeric values like 0,1,2 but then model will see compare the values .In that case use One Hot encoding instead"""
gender_encoder=LabelEncoder()
data['Gender']=gender_encoder.fit_transform(data['Gender'])
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)#Also drop the unnecessary columns from the data
data
```

```python
geo_encoder=OneHotEncoder(sparse_output=False)

encoded_data=geo_encoder.fit_transform(data[['Geography']])
#encoded_data will return a 2D array so convert it into a panda dataframe
encoded_df=pd.DataFrame(encoded_data,columns=geo_encoder.get_feature_names_out(['Geography']))
print(encoded_data)
encoded_df
```

```python
data=pd.concat([data,encoded_df],axis=1)#axis=1 means along the column
#Now we can drop the geography column from actual dataframe along axis=1 column and inplace=True
data.drop('Geography',axis=1,inplace=True)
data
```

```python
#Now make X_train,X_test,y_train and y_test from the data
from sklearn.model_selection import train_test_split
X=data.drop('Exited',axis=1)
y=data['Exited']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)#test_size is 30%of the data and remaining 70%of data will be used for training
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
#We use scale and fit transform the X_train and just transform the X_test
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_train
```

```python
## Save the encoders and sscaler as pickle files
""""""
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(gender_encoder,file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(geo_encoder,file)

with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
X_train.shape[1]
#This is will be number of input parameters for one datapoint one data point will have 12 input features and there are 7000 such datapoints
```

```python
#Model
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),#The first hidden layerconnected to input layer collecting 12 inputs
    Dense(32,activation='relu'),#The second hiddlen layer connected to first hidden layer
    Dense(1,activation='sigmoid')#The output layer
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
#Here we selected the optimizer and loss function for calculating loss and use that in updation of weights
```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
#Set up the Tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## Set up Early Stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
### Train the model
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tensorboard_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
